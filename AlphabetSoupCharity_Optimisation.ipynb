```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("https://static.bc-edx.com/data/dla-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
new_app_df = application_df.drop(columns=["EIN", "NAME"])
new_app_df.head()
```

```python
# Determine the number of unique values in each column.
count = new_app_df.nunique()
print(count)
```

```python
# Look at APPLICATION_TYPE value counts for binning
count_type = new_app_df.APPLICATION_TYPE.value_counts()
print(count_type)
```

```python
# Choose a cutoff value(Higher than last time) and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_to_replace = list(count_type[count_type<1000].index)

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
count_class = new_app_df.CLASSIFICATION.value_counts()
print(count_class)
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
new_app_df['CLASSIFICATION'].value_counts().loc[lambda x : x >1]
```

```python
# Choose a cutoff value(Higher than standard_code) and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace = list(count_class[count_class<1000].index)

# Replace in dataframe
for cls in classifications_to_replace:
    new_app_df['CLASSIFICATION'] = new_app_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
new_app_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
app_dummies = pd.get_dummies(new_app_df)
app_dummies.head(5)
```

```python
# Split our preprocessed data into our features and target arrays
y = app_dummies.IS_SUCCESSFUL.values
X = app_dummies.drop(columns="IS_SUCCESSFUL").values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

Attempt #1 At Optimising (Higher First hidden Layer)

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
input_features = X_train_scaled.shape[1]

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=1600, activation="relu", input_dim= input_features))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=30, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
